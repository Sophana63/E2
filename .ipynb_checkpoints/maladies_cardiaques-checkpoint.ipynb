## Prédiction des maladies cardiaques avec le Machine Learning

Les maladies cardiaques décrivent une gamme de conditions qui affectent votre cœur. Les maladies sous l'égide des maladies cardiaques comprennent les maladies des vaisseaux sanguins, telles que les maladies coronariennes, les problèmes de rythme cardiaque (arythmie) et les malformations cardiaques avec lesquelles vous êtes né (malformations cardiaques congénitales), entre autres.

Les maladies cardiaques sont l'une des principales causes de morbidité et de mortalité parmi la population mondiale. La prédiction des maladies cardiovasculaires est considérée comme l'un des sujets les plus importants dans la section de la science des données cliniques. La quantité de données dans le secteur de la santé est énorme.

Dans ce projet de science des données, j'appliquerai des techniques d'apprentissage automatique pour classer si une personne souffre ou non d'une maladie cardiaque.

```python
# Importation des modules requis

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import rcParams
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

from sklearn.neighbors import KNeighborsClassifier
```

```python
df = pd.read_csv('heart.csv')
print(df.head())
print(df.info())
print(df.describe())
```

```python
# corrélation de chaque caractéristique dans l'ensemble de données

corrmat = df.corr()
top_corr_features = corrmat.index
plt.figure(figsize=(16,16))
#plot heat map
g=sns.heatmap(df[top_corr_features].corr(),annot=True,cmap="RdYlGn")
plt.show()
```

C'est toujours une bonne pratique de travailler avec un ensemble de données où les classes cibles sont de taille approximativement égale

```python
sns.set_style('whitegrid')
sns.countplot(x='target',data=df,palette='RdBu_r')
plt.show()
```

### Traitement de l'information

Après avoir exploré l'ensemble de données, j'ai observé que je devais convertir certaines variables catégorielles en variables fictives et mettre à l'échelle toutes les valeurs avant de former les modèles d'apprentissage automatique.

Tout d'abord, je vais utiliser la méthode get_dummies pour créer des colonnes factices pour les variables catégorielles.

```python
dataset = pd.get_dummies(df, columns = ['sex', 'cp', 
                                        'fbs','restecg', 
                                        'exang', 'slope', 
                                        'ca', 'thal'])
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
standardScaler = StandardScaler()
columns_to_scale = ['age', 'trestbps', 'chol', 'thalach', 'oldpeak']
dataset[columns_to_scale] = standardScaler.fit_transform(dataset[columns_to_scale])
dataset.head()
```

```python
y = dataset['target']
X = dataset.drop(['target'], axis = 1)

from sklearn.model_selection import cross_val_score
knn_scores = []
for k in range(1,21):
    knn_classifier = KNeighborsClassifier(n_neighbors = k)
    score=cross_val_score(knn_classifier,X,y,cv=10)
    knn_scores.append(score.mean())

plt.figure(figsize=(16,16))
plt.plot([k for k in range(1, 21)], knn_scores, color = 'red')
for i in range(1,21):
    plt.text(i, knn_scores[i-1], (i, knn_scores[i-1]))
plt.xticks([i for i in range(1, 21)])
plt.xlabel('Number of Neighbors (K)')
plt.ylabel('Scores')
plt.title('K Neighbors Classifier scores for different K values')
plt.show()
```

```python
knn_classifier = KNeighborsClassifier(n_neighbors = 12)
score=cross_val_score(knn_classifier,X,y,cv=10)
score.mean()
```

```python
from sklearn.ensemble import RandomForestClassifier
randomforest_classifier= RandomForestClassifier(n_estimators=10)
score=cross_val_score(randomforest_classifier,X,y,cv=10)
score.mean()
```
